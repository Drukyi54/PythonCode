##Import Libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

##Ingest data from csv

```python
df=pd.read_csv("/content/data_news.csv")
```

```python
df.head()
```

##Manage differnet datatypes and data wrangling

Drop unnamed files

```python
df.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
df.info()
```

```python
df.describe()
```

Drop the NaN column names "desc" and "site"

```python
columns = ['desc', 'site']
df.drop(columns=columns, inplace=True)
```

```python
df.head()
```

```python
print(df.isnull().sum())
```

```python
df_cleaned = df.dropna()
df_cleaned
```

##visualize data

Below is the visualization of news published by different media

```python
plt.figure(figsize=(18, 8))
sns.countplot(data=df_cleaned, x='media')
plt.title('Distribution of Articles by Media')
plt.xlabel('Media')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.show()
```
